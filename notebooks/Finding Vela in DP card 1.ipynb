```python
nant = 30
idx = 0
products = []
revproducts = {}
auto_products = []
for a1 in range(1, nant+1):
    for a2 in range(a1, nant+1):
        print(f'{a1}-{a2}:{idx}')
        products.append((a1,a2))
        revproducts[(a1,a2)] = idx
        if a1 == a2:
            auto_products.append(idx)
            
        idx += 1

```

```python
revproducts[(1,4)] 
```

```python
autos = list(filter(lambda bl: bl[0] == bl[1],products))
```
