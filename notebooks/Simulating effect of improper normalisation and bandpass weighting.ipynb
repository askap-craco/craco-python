```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(0, 1, (100, 10000))
x[:, 50] = 10
```

```python
siny = (np.sin(np.arange(100)/1.55) /10 + 1)
expy = np.exp(-1*np.arange(100)/100)
expy += 1 - np.mean(expy)
```

```python
plt.figure()
plt.plot((expy + siny)/2)
```

```python
x2 = x.copy()
for ii in range(len(x)):
    x2[ii] = x[ii] * (expy[ii] + siny[ii])/2
```

```python
np.std(x2.ravel()), np.std(x.ravel())
```

```python
plt.figure()
plt.plot(np.std(x, axis=-1))
plt.plot(np.std(x2, axis=-1))
```

```python
plt.figure()
plt.imshow(x2[:, :500], aspect='auto')
```

```python
ts = x.sum(axis=0)
ts2 = x2.sum(axis=0)
```
